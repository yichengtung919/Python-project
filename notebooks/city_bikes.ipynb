# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import numpy as np
from scipy import stats
```

```python
city_bikes_API = 'http://api.citybik.es/v2/networks'
response = requests.get(city_bikes_API)

```

```python
if response.status_code == 200:
    # Request was successful
    try:
        data = response.json()  # Convert response to JSON format
        print(data)  # Print the retrieved data
    except Exception as e:
        print('Error parsing JSON:', e)
        print('Response text:', response.text)  
else:
    # Request failed
    print('Error:', response.status_code)
```

```python
data
```

```python
data2 = data['networks']
data2
```

```python
filtered_data = [entry['href']
                 for entry in data2 if entry['location']['city'] == 'Vancouver']
filtered_data
# obtain city id for Vancouver
```

```python
#use the city id in the previous code to filter the outcome of the request
city_bikes_api_vancouver = 'http://api.citybik.es/v2/networks/mobibikes'
response = requests.get(city_bikes_api_vancouver)
```

```python
if response.status_code == 200:
    # Request was successful
    try:
        data = response.json()  # Convert response to JSON format
        print(data)  # Print the retrieved data
    except Exception as e:
        print('Error parsing JSON:', e)
        print('Response text:', response.text)  # Print the response text for debugging
else:
    # Request failed
    print('Error:', response.status_code)
    
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
vancouver_bike_data = data['network']['stations']
vancouver_bike_data
```

```python
vancouver_bike_data_filtered = [{'id':entry['id'],
                    'name':entry['name'],
                    'latitude': entry['latitude'],
                  'longitude': entry['longitude'],
                  'available_bikes' : entry['free_bikes'],
                  'unavailable_bikes' : entry ['empty_slots'],
                  }
                 for entry in vancouver_bike_data]
vancouver_bike_data_filtered
```

Put your parsed results into a DataFrame.

```python
import pandas as pd
```

```python
df = pd.DataFrame(vancouver_bike_data_filtered)

df
```

```python
z_scores_available = np.abs(stats.zscore(df['available_bikes']))

df[(z_scores_available<3)]
#check for outliers and 2 found
```

```python
z_scores_unavailable = np.abs(stats.zscore(df['unavailable_bikes']))

df[(z_scores_unavailable<3)]
#check for outliers and 1 found
```

```python
df[(z_scores_unavailable<3) & (z_scores_available<3)]
#There are two outliers from the available bikes column and one from the unavailable bikes column but I will include them in the dataset because they are the actual data and 
#it make sense that some of the bike station may contain more bikes than the other.
```
